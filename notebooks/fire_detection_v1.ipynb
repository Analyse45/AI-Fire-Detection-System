### Fire Detection

- Version 1.
   - Project Overview

The Fire Exigency Detection and Emergency Alert System is a real-time computer vision–based safety monitoring application developed using Python and OpenCV. The primary objective of the system is to detect potential fire-like regions from a live webcam feed and trigger immediate emergency responses through alarms, evidence logging, and automated Telegram notifications.

This project was initially conceptualized as a traditional computer vision prototype during the bachelor’s program and later rebuilt with improved engineering practices, modular logic, and real-time monitoring capabilities.

The system continuously analyzes live video frames, identifies fire-colored regions using HSV color segmentation, detects suspicious regions through contour analysis, classifies severity levels, and automatically performs alert actions such as screenshot capture, CSV event logging, sound alarms, and Telegram-based emergency notifications.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import datetime
import os
import time
import winsound
import requests
import csv

# -----------------------------------
# CREATE REQUIRED FOLDERS
# -----------------------------------

os.makedirs("../outputs", exist_ok=True)
os.makedirs("../screenshots", exist_ok=True)

print("Folders ready ✅")

# -----------------------------------
# CREATE CSV LOG FILE
# -----------------------------------

log_file = "../outputs/fire_logs.csv"

if not os.path.exists(log_file):

    with open(log_file, mode='w', newline='') as file:

        writer = csv.writer(file)

        writer.writerow([
            "Timestamp",
            "Severity",
            "Screenshot"
        ])

print("CSV logging system ready ✅")

# -----------------------------------
# FIRE HSV RANGE
# -----------------------------------

lower_fire = np.array([5, 50, 50], dtype="uint8")
upper_fire = np.array([35, 255, 255], dtype="uint8")

print("HSV fire range initialized ✅")

# -----------------------------------
# TELEGRAM CONFIG
# -----------------------------------

BOT_TOKEN = "PASTE_YOUR_BOT_TOKEN_HERE"

CHAT_ID = "PASTE_YOUR_CHAT_ID_HERE"

print("Telegram configuration loaded ✅")

# -----------------------------------
# SEND TELEGRAM ALERT FUNCTION
# -----------------------------------

def send_telegram_alert(message):

    url = f"https://api.telegram.org/bot{BOT_TOKEN}/sendMessage"

    payload = {
        "chat_id": CHAT_ID,
        "text": message
    }

    try:

        response = requests.post(url, data=payload)

        if response.status_code == 200:
            print("Telegram alert sent ✅")

        else:
            print("Failed to send Telegram alert ❌")

    except Exception as e:

        print("Telegram Error:", e)

# -----------------------------------
# START WEBCAM
# -----------------------------------

video = cv2.VideoCapture(0)

if not video.isOpened():
    print("Webcam access failed ❌")
else:
    print("Webcam initialized successfully ✅")

print("Press Q to exit")

# cooldown settings to prevent alert spamming
last_capture_time = 0
cooldown_seconds = 5

# main loop for real-time fire detection
while video.isOpened():

    # Read frame
    ret, frame = video.read()

    if not ret:
        print("Failed to capture frame ❌")
        break

    # Resize frame
    frame = cv2.resize(frame, (800, 500))

    # Convert BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Create fire mask
    mask = cv2.inRange(hsv, lower_fire, upper_fire)

    # Find contours
    contours, _ = cv2.findContours(
        mask,
        cv2.RETR_TREE,
        cv2.CHAIN_APPROX_SIMPLE
    )

    # Detection logic
    for contour in contours:

        area = cv2.contourArea(contour)

        # Ignore tiny regions
        if area > 1000:

            # Determine severity based on area
            if area > 15000:
                severity = "HIGH"

            elif area > 7000:
                severity = "MEDIUM"

            else:
                severity = "LOW"

            # Get bounding box
            x, y, w, h = cv2.boundingRect(contour)

            # Draw rectangle
            cv2.rectangle(
                frame,
                (x, y),
                (x + w, y + h),
                (0, 0, 255),
                3
            )

            # Detection label
            cv2.putText(
                frame,
                f"FIRE DETECTED! [{severity}]",
                (x, y - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (0, 0, 255),
                2
            )

            # Current time
            current_time = time.time()

            # Cooldown logic
            if current_time - last_capture_time > cooldown_seconds:

                # Alarm beep
                winsound.Beep(1000, 500)

                # Timestamp
                timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

                # Screenshot filename
                filename = f"../screenshots/fire_{timestamp}.jpg"

                # Save screenshot
                cv2.imwrite(filename, frame)

                print(f"Screenshot saved: {filename}")

                # Telegram alert message

                alert_message = f"""
🔥 FIRE DETECTED!

🔥 Severity: {severity}

📸 Screenshot Saved
⏰ Time: {timestamp}

⚠ Emergency Attention Required
"""

                # Send Telegram alert
                send_telegram_alert(alert_message)

                print("Alert function executed ✅")

               # save event to CSV

                with open(log_file, mode='a', newline='') as file:

                    writer = csv.writer(file)

                    writer.writerow([
                        timestamp,
                        severity,
                        filename
                    ])

                print("Detection logged to CSV ✅")

                # Update cooldown timer
                last_capture_time = current_time

    # display mode
    cv2.imshow("Fire Detection System", frame)
    cv2.imshow("Fire Mask", mask)

    # Exit on Q
    key = cv2.waitKey(1)

    if key == ord('q'):
        print("Exiting system...")
        break

# cleanup

video.release()
cv2.destroyAllWindows()

print("System closed safely ✅")
```

### System Workflow
- Step 1 — Live Webcam Monitoring

The system captures real-time video feed using OpenCV’s webcam interface.

- Step 2 — HSV Color Space Conversion

Frames are converted from BGR to HSV color space for better color segmentation and illumination handling.

- Step 3 — Fire Color Segmentation

HSV masking is used to isolate fire-like color ranges such as orange, yellow, and red tones.

- Step 4 — Contour Detection

Contours are extracted from masked regions to identify significant fire-like objects.

- Step 5 — Severity Classification

Based on contour area:

   - LOW
   - MEDIUM
   - HIGH

severity levels are dynamically assigned.

- Step 6 — Emergency Actions

Upon detection:

Bounding boxes are drawn
Alarm sound is triggered
Screenshot evidence is captured
Telegram alert is sent
Event is logged into CSV
